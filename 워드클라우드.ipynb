```python
from wordcloud import WordCloud
from string import digits
import re
```

```python
text=""
with open("data/카톡.txt","r",encoding="utf-8") as f:
    lines = f.readlines()
    for line in lines:
        text += line.replace("ㅋ","").replace("ㅎ", "").replace("사진\n","").replace("이모티콘\n","").replace("ㅅㅂ","").replace("ㅄ","").replace("니엄마","").replace("삭제된 메시지입니다.","").replace("지랄","").replace("오전","").replace("오후","").replace("준형","").replace("종현","").replace("정연택선생님","").replace("민석","").replace("종연","").replace("수혁이","").replace("손재영","").replace("세민","").replace("유승현","").replace("준원","").replace("홍영준","").replace("월","").replace("일","").replace("내","").replace("지금","").replace("나도","").replace("오늘","").replace("잘","").replace("황성대","").replace("년","")
text=re.sub(r'[0-9]+',"",text)
print(text)
```

```python
font_path = "C:/Windows/Fonts/malgun.ttf"
```

```python
wc = WordCloud(font_path=font_path, background_color="white", width=600, height=400)
wc.generate(text)
wc.to_file("result_review.png")
```
